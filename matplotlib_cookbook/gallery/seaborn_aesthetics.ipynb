# <font color="red"><h4 align="center">Seaborn 外观设置</h4></font>

Matplotlib 自定义能力很强，但是却不知道从哪儿开始设置，让图表变得美观。Seaborn 自带主题，而且包括 high-level 的接口用于调整 matplotlib figure 外观。

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, "aesthetics")))

%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
# 默认 MATLAB 风格
sinplot()
```

切换到 seaborn 的默认视图，导入seaborn 包即可

```python
import seaborn as sns
```

```python
sinplot()
sns.plt.show()
```

Seaborn 的默认风格不再是MATLAB风格，而是浅灰色背景、白色方格线。孰好孰坏，看个人口味了。

Seaborn 将 matplotlib 的参数分隔为两部分。第一部分设置 plot的风格；第二部分则是 figure 的不同内容设置。

设置风格：**`axes_style()`** 和 **`set_style`**

设置 plot: **`plotting_context()`** 和 **`set_context()`**

上面两个，第一个方法均返回参数的 dict，第二个设置 matplotlib 的默认值。

## <font color="red">1. 通过 axes_style() 和 set_style() 设置 figure 风格</font>

Seaborn 有5个预定义 themes: darkgrid, whitegrid, dark, white 和 ticks.

默认为 darkgrid.grid 有助于查看定量信息，而灰底白线可以避免和数据点冲突。

```python
sns.set_style('whitegrid')
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
# white 风格
sns.set_style('white')
sinplot()
```

```python
sns.set_style('white')
```
