```python
ax.legend(['curve1', 'curve2', 'curve3'])
```

上面的方法是MATLAB风格API，很容易出错。

推荐在创建 plot 时使用 label="label text" 关键字参数，然后调用不带参数的 `legend` 方法添加 legend:

```python
ax.plot(x, x**2, label="curve1")
ax.plot(x, x**3, label="curve2")
ax.legend()
```

这个方法的优点是：当添加或删除 figure时，对应的 legend 会自动更新。

`legend` 函数有一个可选关键字参数 `loc`，用于指定绘制 legend 的位置。具体参数的含义可以参考 http://matplotlib.org/users/legend_guide.html#legend-location ，常见的参数如下：

```python
ax.legend(loc=0) # let matplotlib decide the optimal location
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner
# .. many more options are available
```

下面演示如何使用 figure title, axis labels and legends:

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label="y = x**2")
ax.plot(x, x**3, label="y = x**3")
ax.legend(loc=2) ## upper left corner
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

### 格式化文本：LaTex, fontsize, font family

发表文章的图表，不仅要求功能，还要求美观，所以，字体调整是必须的。

Matplotlib 对 LaTeX 具有很好的支持。只需要将 LaTeX 内容用美元符括起来。例如，`"$y = x^3$"`.

由于 LaTeX 中大量使用了斜杠，和 python 的转义会干扰，所以请使用 "raw" 字符串。例如 `r"\alpha" `: 

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label=r'$y = \alpha^2$')
ax.plot(x, x**3, label=r'$y = \alpha^3$')
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$', fontsize=18)
ax.set_ylabel(r'$y$', fontsize=18)
ax.set_title('title')
```

也可以修改全局的字体，应用于图表的所有文本：

```python
# Update the matplotlib configuration parameters:
matplotlib.rcParams.update({'font.size':18, 'font.family':'serif'})
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('title');
```

STIX 字体是一个比较合适的全局字体：

```python
# Update the matplotlib configuration parameters:
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('title')
```
